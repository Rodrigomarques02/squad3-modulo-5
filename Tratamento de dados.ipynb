# Exportação para Banco de Dados SQL

```python
def dataframe_to_sql(dataframe, sql_filename, table_name):
    with open(sql_filename, 'w') as arquivo_sql:
        column_names = ', '.join(dataframe.columns)

        arquivo_sql.write(f"INSERT INTO {table_name} ({column_names})\nVALUES\n")
        
        for _, row in dataframe.iterrows():
            values = []

            for value in row:
                if isinstance(value, str):
                    # Adicione aspas simples ao redor de valores de string
                    values.append(f"'{value}'")
                else:
                    values.append(str(value))

            values = ', '.join(values)
            arquivo_sql.write(f"({values}),\n")
        
        # Remova a última vírgula e nova linha e adicione um ponto e vírgula
        arquivo_sql.seek(arquivo_sql.tell() - 2)
        arquivo_sql.truncate()
        arquivo_sql.write(";")
```

```python
if __name__ == "__main__":
    data = data_usdc  # DataFrame
    sql_filename = 'usdcoin.sql'  # Nome do arquivo SQL
    table_name = 'coin_usdc'  # Nome da tabela

    dataframe_to_sql(data, sql_filename, table_name)
```

Essa função cria todos os arquivos de inserts para povoar as tabelas criadas no banco de dados apenas modificando as variáveis ``data``, ``sql_filename`` e ``table_name`` para o dataframe a ser convertido em sql.
